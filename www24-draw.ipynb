```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import numpy as np
from matplotlib import font_manager  # 字体

plt.rcParams['font.sans-serif']=['KaiTi']
#对比两天内同一时刻温度的变化情况
x = [1, 2, 3, 4]
y1 = [86.18, 88.58, 85.35, 84.06]
y2 = [82.84, 86.03, 81.40, 80.17]
#绘制折线图，添加数据点，设置点的大小
# * 表示绘制五角星；此处也可以不设置线条颜色，matplotlib会自动为线条添加不同的颜色



fig, ax = plt.subplots()
ax.plot(x, y1, color='#55b7e6',marker='o', markersize=5, label='准确率（%）',LineWidth=2)
ax.legend(loc=(4/30, 12/16))
ax.plot(x, y2, color='#ed6e69', marker='^',markersize=5, label='F1分数（%）',LineWidth=2)
ax.legend(loc=(16/30, 12/16), prop={'size': 14})
# plt.title('温度对比折线图')  # 折线图标题
ax.set_xlabel('GCN层数', size=14)  # x轴标题
ax.set_ylabel('评价指标（%）', size=14)  # x轴标题
#ax.set_xlim(0.5, 4.5)
ax.set_ylim(78, 90)
plt.xticks(size=14)  #x轴字体
plt.yticks(size=14)
x_major_locator=MultipleLocator(1)
y_major_locator=MultipleLocator(2)
ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(y_major_locator)
#ax.grid(True, axis='y')   # 网格背景
for a, b in zip(x, y1):
    plt.text(a, b, (b),ha='center', va='bottom', fontsize=10)
for a, b in zip(x, y2):
    plt.text(a, b, (b), ha='center', va='bottom', fontsize=10)
plt.savefig("MMSD_GCN_Layers.jpg", dpi=600 )
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']=['KaiTi']
size = 2
x = np.array([0, 1])
a = np.array([86.97, 83.81])
b = np.array([88.58, 86.03])
c = np.array([85.97, 82.57])
d = np.array([86.34, 83.17])

total_width, n = 0.6, 4
width = total_width / n

# plt.grid(True, axis='y', alpha=0.2)   # 网格背景
plt.bar(x, c, width=width, label='-Glove+Resnet', color='#fccccb')
plt.bar(x+width, d, width=width, label="-Glove+ViT", color='#bdb5ea')
plt.bar(x+2*width, a,  width=width, label='-BERT+Resnet', color='#b0d992')
plt.bar(x+3*width, b, width=width, label='-GE-REN', color='#f9d580')
tick_label=["准确率（%）", "F1分数（%）"]
plt.ylim(80, 90)
plt.legend(prop={'size': 12})
plt.xticks(x+width*1.5, tick_label, size=14)#显示x坐标轴的标签,即tick_label,调整位置，使其落在两个直方图中间位置
plt.yticks(size=14)
plt.xlabel('编码方式', size=14)  # x轴标题
plt.ylabel('评价指标（%）', size=14)  # y轴标题
for m, n in zip(x, a):
    plt.text(m+2*width, n, (n),ha='center', va='bottom', fontsize=10)
for m, n in zip(x, b):
    plt.text(m+3*width, n, (n),ha='center', va='bottom', fontsize=10)
for m, n in zip(x, c):
    plt.text(m, n, (n),ha='center', va='bottom', fontsize=10)
for m, n in zip(x, d):
    plt.text(m+width, n, (n),ha='center', va='bottom', fontsize=10)
plt.savefig("MMSD_Diffcompoent.jpg", dpi=300 )
```

```python
# 参数K的分析

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import numpy as np
from matplotlib import font_manager  # 字体

plt.rcParams['font.sans-serif']=['KaiTi']
#对比两天内同一时刻温度的变化情况
x = [0, 5, 10, 15, 20]
y1 = [86.34, 88.58, 84.68, 82.07, 79.99]
y2 = [83.10, 86.03, 82.92, 79.64, 78.32]
#绘制折线图，添加数据点，设置点的大小
# * 表示绘制五角星；此处也可以不设置线条颜色，matplotlib会自动为线条添加不同的颜色



fig, ax = plt.subplots()
ax.plot(x, y1, color='#df2089',marker='o', markersize=5, label='准确率（%）',LineWidth=2)
ax.legend(loc=(4/30, 12/16))
ax.plot(x, y2, color='#20df76', marker='^',markersize=5, label='F1分数（%）',LineWidth=2)
ax.legend(loc=(16/30, 12/16), prop={'size': 14})
# plt.title('温度对比折线图')  # 折线图标题
ax.set_xlabel('$k$的不同取值', size=14)  # x轴标题
ax.set_ylabel('评价指标（%）', size=14)  # x轴标题
ax.set_xlim(-1, 22)
ax.set_ylim(75, 90)

plt.xticks(size=14)  #x轴字体
plt.yticks(size=14)
x_major_locator=MultipleLocator(5)
y_major_locator=MultipleLocator(2)
ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(y_major_locator)
ax.grid(True, axis='x',linestyle='--',alpha=0.5)   # 网格背景
plt.savefig("differ_k_GEREN.jpg", dpi=600 )
```
